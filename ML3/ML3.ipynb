## Part 1 Data Preprocessing
Replace all the nan values in "PH" feature using "sulphates" feature. 

```python
import pandas as pd
import numpy as np
import math

df = pd.read_csv('JuiceQuality.csv')
for item in df['sulphates']:
    print(item)
```

```python
# generate a map for replacing each item with its mean in their 'pH' attribute
map = {
    'low': np.mean(df[df['sulphates'] == 'low']['pH']),
    'medium': np.mean(df[df['sulphates'] == 'medium']['pH']),
    'high': np.mean(df[df['sulphates'] == 'high']['pH']),
    'very_high': np.mean(df[df['sulphates'] == 'very_high']['pH']),
    'nan': np.mean(df[np.isnan(df['pH']) == False]['pH'])
}

# replace nan items in 'pH' column
df['pH'] = df.apply(lambda i: map[str(i['sulphates'])] if math.isnan(i['pH']) else i['pH'], axis = 1)
display(df)
```

Use Linear Regression to fill out nan values. "A Indx" feature is presumed as a Label and "fixed acidity", "citric acid" and " chlorides" are our features. You can use any of the mentioned features to estimate nan values in Label. 

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

# first we should find the best combination of features we can use. so we use rows 
#with specified features and labels

# drop nan of selected colums. all of dataset should have clear data
data_set = df.dropna(subset=['fixed acidity', 'citric acid', 'chlorides', 'A Indx'])

# we should test all combination of features and select the best one!! :/
features = [['fixed acidity'], ['citric acid'], ['chlorides'], ['fixed acidity', 'citric acid'], ['fixed acidity', 'chlorides'], ['citric acid', 'chlorides'], ['fixed acidity', 'citric acid', 'chlorides']]
errors = np.zeros(len(features))
tmp_i = 0

for feature in features:
    # make learning data ready
    X_learn = data_set.loc[:, feature]
    Y_learn = data_set.loc[:, 'A Indx']
    X_train, X_test, Y_train, Y_test = train_test_split(X_learn, Y_learn, test_size=0.15)
    
    # take model and rain it
    regressor = LinearRegression()
    regressor.fit(X_train, Y_train)
    
    # calculate error and save it
    Y_o = regressor.predict(X_test)
    errors[tmp_i] = np.sum((Y_o - Y_test) ** 2) / len(Y_test)
    tmp_i = tmp_i + 1
    
# select and rebuild the best model    
best = (np.where(errors == min(errors))[0][0])
X_learn = data_set.loc[:, features[best]]
Y_learn = data_set.loc[:, 'A Indx']
# because we do not want to estimate any eror now so we use all available data to make the model
regressor = LinearRegression()
regressor.fit(X_learn, Y_learn)

# because we want to fill "A index" based on three other features, first we should fill nan properties in
# that rows
df[['fixed acidity', 'citric acid', 'chlorides']] = round(df[['fixed acidity', 'citric acid', 'chlorides']].fillna(df[['fixed acidity', 'citric acid', 'chlorides']].mean()), 2)

# now we make prediction using our best model and get rid of nan items in "A index" column 
df['A Indx'] = df.apply(lambda i: round(regressor.predict([[i[j] for j in features[best]]])[0], 2) if math.isnan(i['A Indx']) else i['A Indx'], axis = 1)
print(features[best])
display(df)
```

$$C \in \{0.1, 1\}$$

We are going to very briefly and easily look at how we go about solving optimization problems that we encounter when solving SVM. Here we use the same quadratic programming (use solvers.qp()) present in the theory problem set 3. Verify your theoretical solutions with Python code.

$$min\hspace{3mm} J = (x_1-6)^2+(x_2-6)^2$$

$$s.\hspace{1mm} to \hspace{3mm}2 x_1+4 x_2 \leq 28$$

$$5 x_1 + 5 x_2 \leq 50$$

$$x_1 \leq 8$$

$$x_2 \leq 6$$

$$x_1 \geq 0$$

$$x_2 \geq 0$$

```python
from cvxopt import matrix, solvers
```

```python
P = matrix([[2.0, 0.0], [0.0, 2.0]])
q = matrix([-12.0, -12.0])
G = matrix([[1.0, 1.0, 1.0, 0.0, -1.0, 0.0], [2.0, 1.0, 0.0, 1.0, 0.0, -1.0]])
h = matrix([14.0, 10.0, 8.0, 6.0, 0.0, 0.0])
solution = solvers.qp(P, q, G, h)
print(solution['x'])
```
